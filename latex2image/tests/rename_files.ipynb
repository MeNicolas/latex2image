```python
import os
```

```python
cwd = os.getcwd()
print(cwd)

for count, filename in enumerate(os.listdir()): 
    print(filename)
    src = filename
    if not filename.endswith(".tex"):
        continue
    if "xelatex" in filename:
        dst = "xelatex_%03i.tex" % (count +1)
        os.rename(os.path.join(cwd, src), os.path.join(cwd, "xelatex", dst)) 
    else:
        dst = "lualatex_%03i.tex" % (count +1)
        os.rename(os.path.join(cwd, src), os.path.join(cwd, "lualatex", dst)) 
```
